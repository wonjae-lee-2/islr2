```python
private_yes = df.loc[df['Private'] == 'Yes', 'Outstate']
private_no = df.loc[df['Private'] == 'No', 'Outstate']
data = [private_yes, private_no]
labels = ['Yes', 'No']

fig, ax = plt.subplots()
ax.boxplot(data, labels=labels)
ax.set_xlabel('Private')
ax.set_ylabel('Outstate');
```

```python
df.boxplot(column='Outstate', by='Private');
```

```python
df['Elite'] = np.where(df.Top10perc > 50, 'Yes', 'No')
df.loc[df.Elite == 'Yes', 'Elite']
```

```python
elite_yes = df.loc[df['Elite'] == 'Yes', 'Outstate']
elite_no = df.loc[df['Elite'] == 'No', 'Outstate']
data = [elite_yes, elite_no]
labels = ['Yes', 'No']

fig, ax = plt.subplots()
ax.boxplot(data, labels=labels)
ax.set_xlabel('Elite')
ax.set_ylabel('Outstate');
```

```python
df.boxplot(column='Outstate', by='Elite');
```

```python
data = df.select_dtypes(include='number')

fig, axs = plt.subplots(nrows=4, ncols=4)
fig.set_size_inches(20, 20, forward=False)
for i in range(16):
    axs.flat[i].hist(x=data.iloc[:, i])
    axs.flat[i].set_xlabel(data.iloc[:, i].name)
```
