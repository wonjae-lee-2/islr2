## Exercise 2.9

```python
df = pd.read_csv(p / 'Auto.csv')
df.info()
```

```python
df['horsepower'].unique()
```

```python
df = df[df['horsepower'] != '?']
df = df.astype({'horsepower': 'int64'})
df.info()
```

### Problem (a)

```python
df
```

```python
qual = ['origin', 'name']
quant = [x for x in df.columns if x not in qual]
quant
```

### Problem (b)

```python
df_quant = df[quant].copy()
df_quant.apply(lambda x: x.max() - x.min())
```

### Problem (c)

```python
df_quant.describe().loc[['mean', 'std'], :]
```

### Problem (d)

```python
(
    df_quant
    .reset_index(drop=True)
    .drop(range(9, 84))
    .agg([lambda x: x.max() - x.min(), 'mean', 'std'])
    .rename(index={'<lambda>': 'range'})
)
```
